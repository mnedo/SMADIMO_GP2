```python
wiki_url = 'https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%B3%D0%BE%D1%80%D0%BE%D0%B4%D0%BE%D0%B2_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8'
headers = {'User-Agent': 'project1'}#src про scraping wiki: https://automatio.ai/ru/how-to-scrape/wikipedia
wiki = requests.get(wiki_url, headers=headers)
logger.info(f'Запрос к Wikipedia: статус {wiki}')
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(wiki.text, 'html')
```

```python
table = soup.find('table').find_all('tr')
table[1].find_all('td')
```

Спарсим все города со страницы википедии

```python
cities = []
for elem in table[1:]:
  cities.append(elem.find_all('td')[2].find('a').get('title'))
logger.info(f'парсинг городов, всего: {len(cities)} шт')
```

Также нам необходимо спарсить регионы и федеральные округа

```python
regions = []
for elem in table[1:]:
  regions.append(elem.find_all('td')[3].text)
logger.info(f'парсинг регионов, всего: {len(regions)} шт')
regions
```

```python
federal_districts = []
for elem in table[1:]:
  federal_districts.append(elem.find_all('td')[4].text)
logger.info(f'парсинг округов, всего: {len(regions)} шт')
federal_districts
```

На нашей странице численность, которая была актуальная на момент 2021 года, однако если детально ппереходить на города, то можно найти более свежие данные. Именно поэтому собираем все ссылки на странички городов и оттуда вытягиваем последнюю доступную численность

```python
urls = []
for elem in table[1:]:
  urls.append(elem.find_all('td')[2].find('a').get('href'))
logger.info(f'сбор ссылок на страницы городов: {len(regions)} шт')
urls
```

```python
import time
```

```python
populations = []
k = 0
for url in urls:
  k += 1
  logger.info(f'парсинг городов (индив): {k} / {len(urls)}')
  page = requests.get('https://ru.wikipedia.org' + url, headers=headers)
  soup = BeautifulSoup(page.text, 'html')
  tables = soup.find_all('table')
  for t in tables:
      caption = t.find('caption')
      if caption and 'Численность населения' in caption.text:
          population_table = t
          break
  population_table1 = population_table.find_all('tr')[-1].find_all('td')
  for i in population_table1[::-1]:
      if i.text:
          answ = i
          break
  populations.append(answ.text.replace('\xa0', '')[1:])
  time.sleep(1)
```

Для нашего будущего датсета нам понадобится информация об аэропортах - для этого нам необходимо спарсить соответствующую страничку википедии

```python
wiki_air = requests.get('https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%B0%D1%8D%D1%80%D0%BE%D0%BF%D0%BE%D1%80%D1%82%D0%BE%D0%B2_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8', headers=headers)
logger.info(f'Запрос на страницу аэропортов: статус {wiki_air}')
```

```python
wiki_air
```

```python
soup_air = BeautifulSoup(wiki_air.text, 'html')
```

Создадим датафрейм с теми городами, в которых есть аэропорт

```python
airports_cities = []
airports_regions = []
for air in soup_air.find('table').find_all('tr')[1:]:
  a = air.find_all('td')
  c = a[5].find('a').get('title')
  r = a[6].text.strip()
  if not((c in airports_cities) and (r in airports_regions)):
    airports_cities.append(c)
    airports_regions.append(r)
airports = pd.DataFrame({'Город': airports_cities, 'Регион': airports_regions})
logger.info(f'парсинг аэропортов: {len(airports)} шт')
```

```python
airports['Наличие аэропорта'] = True
```

# Подготовка датасета
### шаг 1
Мы заметили, что при парсинге названий городов некоторые объекты идут с уточнениями, например, "Абаза (город)" - очистим подобные случаи и оставим только полное название городов

```python
data['Город'] = data['Город'].apply(lambda x: x.split(' (')[0])
```

```python
len(data['Город'].unique())
```

### шаг 2
Сделаем признак "Население" числовым

```python
data['Население'] = data['Население'].astype('int')
```

```python
data.info()
```

### шаг 3
Добавить инфо об аэропортах

```python
data2 = data.merge(airports, on=['Город', 'Регион'], how='left')
data2['Наличие аэропорта'].notna().sum()
```
